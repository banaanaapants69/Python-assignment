```
from scipy.optimize import root
from math import cos
def eq(x):
  return (x+cos(x))

r=root(eq,0)
print(r)
```

```
from scipy.optimize import minimize

def eq(x):
  return (x**2)+x+2

min=minimize(eq,0,method="BFGS")
print(min)
```
