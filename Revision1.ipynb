```
# Program for finding LCM & HCF of two input numbers
def computehcf(x,y):
  #choose the smaller number
    if x>y:
      smaller=y
    else:
      smaller=y
    for i in range(1,smaller+1):
      if((x%i==0) and (y%i==0)):
        hcf=i
    return hcf      

num1=int(input("enter the first number\t"))
num2=int(input("enter the second number\t"))
print("the hcf of",num1,"and",num2,"is",computehcf(num1,num2))
```

```
# python program to find the lcm and hcf of two input numbers

def lcm(x,y):

    if x>y:
        greater=x
    else:
        greater=y
    while(True):
        if ((greater%x==0) and (greater%y==0)):
            lcm=greater
            break
        greater+=1
    return lcm
num1=int(input("enter first number\t"))
num2=int(input("enter second number\t"))
print("The LCM of",num1,"and",num2,"is",lcm(num1,num2))
```
